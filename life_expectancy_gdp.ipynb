```python
from matplotlib import pyplot as plt
import pandas as pd
import seaborn as sns


```

```python
df = pd.read_csv('all_data.csv')
df.head()
```

```python
df.shape
```

```python
print(df.Country.unique())
```

```python
print(df.Year.unique())
```

```python
df = df.rename({"Life expectancy at birth (years)":"Life"}, axis = "columns")
df.head()
```

```python
plt.figure(figsize=(8,6))
sns.distplot(df.Life, rug = True, kde=False)
plt.xlabel("Life expectancy at birth (years)");
```

```python
plt.figure(figsize=(8,6))
sns.displot(df.GDP, rug=True, kde=False)
plt.xlabel('GDP in Trillions of U.S. Dollars');
```

```python
dfAvg = df.drop('Year',axis=1).groupby('Country').mean().reset_index()
```

```python
dfAvg
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Country', y='Life', data=dfAvg, hue='Country')
plt.xlabel('Life expectancy at birth (years');
```

```python
plt.figure(figsize=(10,6))
sns.barplot(x='Country', y='GDP', data=dfAvg, hue='Country')
plt.xlabel('GDP in Trillions of Dollars')
```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=df.Year, y=df.GDP, hue=df.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel("GDP in Trillions of U.S. Dollars");


```

```python
plotGDP = sns.FacetGrid(df, col='Country', col_wrap=3,
                        hue ='Country', sharey =False)

plotGDP = (plotGDP.map(sns.lineplot,'Year','GDP')
           .add_legend()
           .set_axis_labels('Year','GDP in Trillions of U.S. DOllars'))
plotGDP;

```

```python
plt.figure(figsize=(8,6))
sns.lineplot(x=df.Year, y=df.Life, hue=df.Country)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.ylabel('Life expectancy at birth (years)');

```

```python
plotLife = sns.FacetGrid(df, col="Country", col_wrap=3,
                      hue = "Country", sharey = False)

plotLife = (plotLife.map(sns.lineplot,"Year","Life")
         .add_legend()
         .set_axis_labels("Year","Life expectancy at birth (years)"))

plotLife;
```

```python
sns.scatterplot(x=df.Life, y=df.GDP, hue=df.Country).legend(loc='center left', bbox_to_anchor=(1, 0.5), ncol=1)
plt.xlabel('Life expectancy at birth (years)');

```

```python
plot = sns.FacetGrid(df, col='Country', col_wrap=3,
                     hue ='Country', sharey = False, sharex = False)
plot = (plot.map(sns.scatterplot,'Life','GDP')
        .add_legend()
        .set_axis_labels('Life expectancy at birth (years)','GDP in Trillions of U.S. Dollars'));


```
